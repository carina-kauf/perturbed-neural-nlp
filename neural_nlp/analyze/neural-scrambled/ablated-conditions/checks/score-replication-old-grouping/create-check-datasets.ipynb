```python
import pickle
import os

from neural_nlp.benchmarks.neural import *
from neural_nlp.models import *
import neural_nlp
from neural_nlp.models.implementations import *
from neural_nlp.stimuli import StimulusSet
import xarray as xr

benchmark_pool = [
    ('Pereira2018-encoding', PereiraEncoding),
    ('Pereira2018-encoding-scrambled-original', PereiraEncodingScrambledOriginal),
]

benchmark_pool = {identifier: LazyLoad(lambda identifier=identifier, ctr=ctr: ctr(identifier=identifier))
                  for identifier, ctr in benchmark_pool}

# fetch stimulus set
benchmark_test = benchmark_pool['Pereira2018-encoding']
stimuli_df = benchmark_test._target_assembly.attrs['stimulus_set']
stimuli_df
```

```python
sentences = list(stimuli_df["sentence"])

import re

no_punctuation_sentences = [re.sub(r'[^\w\s\'\-]','',sentences[i]) for i in range(len(sentences))]
no_punctuation_sentences

# #test functionality
# for i in range(10):
#     print(sentences[i])
#     print(re.sub(r'[^\w\s\'\-]','',sentences[i])) #remove everything except words and space
#     print("\n")
```

```python
no_punctuation_df = stimuli_df.copy()
no_punctuation_df["sentence"] = no_punctuation_sentences

savedir = '/om/user/ckauf/neural-nlp/ressources/scrambled-stimuli-dfs'
fname = f'stimuli_nopunctuation.pkl'
with open(os.path.join(savedir, fname), 'wb') as fout:
    pickle.dump(no_punctuation_df, fout)
no_punctuation_df
```

```python
with open(os.path.join(savedir, "stimuli_lowPMI.pkl"), 'rb') as f:
    lowpmi_df = pickle.load(f)
    
lowpmi_sentences = list(lowpmi_df["sentence"])
lowpmi_finalperiod = [elm + "." for elm in lowpmi_sentences]

lowpmi_finalperiod_df = stimuli_df.copy()
lowpmi_finalperiod_df["sentence"] = lowpmi_sentences

savedir = '/om/user/ckauf/neural-nlp/ressources/scrambled-stimuli-dfs'
fname = f'stimuli_lowpmi_finalperiod.pkl'
with open(os.path.join(savedir, fname), 'wb') as fout:
    pickle.dump(lowpmi_finalperiod_df, fout)
lowpmi_finalperiod_df
```

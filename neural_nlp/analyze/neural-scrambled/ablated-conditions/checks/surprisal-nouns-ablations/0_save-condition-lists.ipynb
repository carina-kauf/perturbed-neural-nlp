```python
import numpy as np

def get_unique_identifiers(df):
    zipped = list(zip(list(df["experiment"]),list(df["passage_index"])))
    unique_tuples = []
    for elm in zipped:
        if elm not in unique_tuples:
            unique_tuples.append(elm)
    print(len(unique_tuples))
    return unique_tuples
unique_tuples = get_unique_identifiers(df)
unique_tuples
```

```python
def get_passage_sentence_list(df, unique_tuples):
    passage_sentences = []

    for elm in unique_tuples:
        curr_df = df.loc[(df["experiment"] == elm[0]) & (df["passage_index"] == elm[1])]
        #print(curr_df)
        curr_sentences = ". ".join(list(curr_df["sentence"]))
        curr_sentences += "."
        passage_sentences.append(curr_sentences)
    print(len(passage_sentences))
    
    return(passage_sentences)
passage_sentences = get_passage_sentence_list(df, unique_tuples)
passage_sentences
```

```python
passage_sentences
```

```python
def get_sentence_lists(picklefile):
    with open(os.path.join(df_dir,picklefile), "rb") as f:
        df = pickle.load(f)
    unique_tuples = get_unique_identifiers(df)
    passage_sentences = get_passage_sentence_list(df, unique_tuples)
    
    #print(passage_sentences[:3])
    
    condition = picklefile.split(".")[0].split("_")[1]
    savename = f"sentence_file_{condition}_bypassages.txt"
    #print(savename)
    #print(os.getcwd())
    
    with open(savename, 'w') as f:
        for item in passage_sentences:
            f.write("%s\n" % item)
    
    return passage_sentences
```

```python
target_conditions = ["stimuli_Original.pkl", "stimuli_nouns.pkl",
                     "stimuli_nounsverbs.pkl", "stimuli_nounsverbsadj.pkl",
                    "stimuli_contentwords.pkl"]
```

```python
def main():
    for cond in target_conditions:
        print(cond)
        get_sentence_lists(os.path.join(df_dir,cond))
    print("Done")
main()
```

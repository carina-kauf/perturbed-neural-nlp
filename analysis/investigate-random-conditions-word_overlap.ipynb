```python
with open(os.path.join(savedir, "stimuli_Original.pkl"), "rb") as f:
    df_original = pickle.load(f)
original_sentences = list(df_original["sentence"])
df_original.head()
```

```python
with open(os.path.join(savedir, "stimuli_random.pkl"), "rb") as f:
    df_random = pickle.load(f)
random_sentences = list(df_random["sentence"])
df_random.head()
```

```python
with open(os.path.join(savedir, "stimuli_random_poscontrolled.pkl"), "rb") as f:
    df_random_pos = pickle.load(f)
random_sentences_pos = list(df_random_pos["sentence"])
df_random_pos.head()
```

```python
def get_overlap_stats(original_sentences, random_sentence_list):
    ratios_overlap = []
    lengths = []
    for ind, elm in enumerate(original_sentences):
        words = elm.split()
        random_words = random_sentence_list[ind].split()
        length = len(words)
        count = sum([1 for x in random_words if x in words])
        curr_ratio = count/length
        ratios_overlap.append(curr_ratio)
        lengths.append(length)
        if ind%50 == 0:
            print(f"{ind} | {elm} | {random_sentence_list[ind]}")
            print(f"Overlapping words: {[x for x in random_words if x in words]}")
            print(f"Overlap count: {count} | Current length: {length} | Overlap ratio: {curr_ratio}")
            print("\n")
    print(f"Average overlap percentage: {np.mean(ratios_overlap)}")
    print(f"Average number of words that overlap: {np.mean(ratios_overlap)/np.mean(lengths)}")
    print(f"Average number of words per sentence: {np.mean(lengths)}")
```

# Compare the two random wordlist conditions to the original stimuli

```python
for ind, sentences in enumerate([random_sentences, random_sentences_pos]):
    if ind == 0:
        print("random-wordlist \n")
    else:
        print("random-wordlist | POS controlled \n")
    get_overlap_stats(original_sentences, sentences)
    print("\n*******\n")
```

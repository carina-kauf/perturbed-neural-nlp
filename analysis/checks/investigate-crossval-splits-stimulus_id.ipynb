```python
def prep_content(filename):
    with open(os.path.join(working_dir,filename_orig), "r") as f:
        file = f.read()
    content = file.split("splits in Split | ")[1].split("\nindices in Split")[0]
    content = re.sub("\n\\s+", " ", content)
    content = content.split("array(")[1:]
    content = [re.sub("\)\), \('","", elm) for elm in content]
    content = [re.sub("\]\), ","]", elm) for elm in content]
    content = [re.sub("\]\)\), \(", "]", elm) for elm in content]
    content = [re.sub("\]\)\)\]", "]", elm) for elm in content]
    content = [re.sub("\s", "" , elm) for elm in content]
    content = [re.sub("\[", "" , elm) for elm in content]
    content = [re.sub("\]", "" , elm) for elm in content]
    content
    
    my_list = []
    for elm in content:
        curr = elm.split(",")
        curr = [int(i) for i in curr]
        my_list.append(curr)
    print(len(my_list))
    my_list
    
    train_splits = [my_list[ind] for ind in [0,2,4,6,8]]
    test_splits = [my_list[ind] for ind in [1,3,5,7,9]]
    
    zipped = list(zip(train_splits,test_splits))
    
    for ind, (train_iter,test_iter) in enumerate(zipped):
        assert all(x not in train_iter for x in test_iter)
    print("Asserted!")
    return train_splits, test_splits
```

```python
train_split_passage_index, test_split_passage_index = prep_content(filename_passage)
train_split_stimulus_id, test_split_stimulus_id = prep_content(filename_orig)
train_split_passage_category, test_split_passage_category = prep_content(filename_topic)
```

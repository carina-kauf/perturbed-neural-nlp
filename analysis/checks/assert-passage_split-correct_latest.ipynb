```python
from neural_nlp.benchmarks.neural import *
import neural_nlp
from neural_nlp.stimuli import StimulusSet
import xarray as xr

benchmark_pool = [
    # primary benchmarks
    ('Pereira2018-encoding', PereiraEncoding),
]
benchmark_pool = {identifier: LazyLoad(lambda identifier=identifier, ctr=ctr: ctr(identifier=identifier))
                  for identifier, ctr in benchmark_pool}

# fetch stimulus set
benchmark = benchmark_pool['Pereira2018-encoding']
stimuli_df = benchmark._target_assembly.attrs['stimulus_set']
stimuli_df
```

```python
curr_df = stimuli_df.loc[stimuli_df["experiment"]=="384sentences"]
curr_df
```

```python
unique_passage = np.unique(curr_df["passage_index"])
print(len(unique_passage))
```

```python
passages = list(curr_df["passage_index"])
sentence_ids = list(curr_df["sentence_num"])
passage_sent = list(zip(passages,sentence_ids))
```

```python
passage_dict = {}
for ind, (passage, sent_id) in enumerate(passage_sent):
    if passage not in passage_dict:
        passage_dict[passage] = [sent_id]
    else:
        passage_dict[passage].append(sent_id)
passage_dict
```

```python
with open("passagesplit_formatted.txt", "r") as f:
    content = f.readlines()
content
```

```python
TRAIN = [elm for elm in content if elm.startswith("TRAIN")]
TEST = [elm for elm in content if elm.startswith("TEST")]
```

```python
TRAIN = [elm.lstrip("TRAIN:[").rstrip("]\n") for elm in TRAIN]
TEST = [elm.lstrip("TEST:[").rstrip("]\n") for elm in TEST]
```

```python
TEST
```

```python
train_splits = []
for elm in TRAIN:
    current_split = [int(i) for i in elm.split(",")]
    train_splits.append(current_split)
    
test_splits = []
for elm in TEST:
    current_split = [int(i) for i in elm.split(",")]
    test_splits.append(current_split)
```

```python
test_splits
```

```python
# assert non-overlap between each train & test split
zipped = list(zip(train_splits,test_splits))
for ind, (train_iter,test_iter) in enumerate(zipped):
    assert all(x not in train_iter for x in test_iter)
print("Asserted!")
```

```python
# assert non-overlap between test splits
for i in range(len(test_splits)-1):
    assert all(x not in test_splits[i+1] for x in test_splits[i])
print("Asserted!")
```

```python
print_file = "investigate-crossval-splits_print-output_passage-split_latest.txt"
```

```python
with open(print_file,'w') as f:
    for ind, (train_iter,test_iter) in enumerate(zipped):
        print(ind)
        for key,value in passage_dict.items():
            #assert all(x in train_iter for x in value) or all(x in test_iter for x in value)
            if all(x in train_iter for x in value):
                print(f"All sentences of passage *{key}* are in the TRAIN set for split {ind}: {value}!")
                f.write(f"All sentences of passage *{key}* are in the TRAIN set for split {ind}: {value}!\n")
            elif all(x in test_iter for x in value):
                print(f"All sentences of passage *{key}* are in the TEST set for split {ind}: {value}!")
                f.write(f"All sentences of passage *{key}* are in the TEST set for split {ind}: {value}!\n")
            else:
                print(f"ERROR! Some sentences from passage *{key}* are in TEST, some in TRAIN for split {ind}:")
                f.write(f"ERROR! Some sentences from passage *{key}* are in TEST, some in TRAIN for split {ind}:\n")
                elms_in_train = []
                elms_in_test = []
                for elm in value:
                    if elm in train_iter:
                        elms_in_train.append(elm)
                    elif elm in test_iter:
                        elms_in_test.append(elm)
                    else:
                        raise ValueError(f"{elm} neither in train nor test!")
                print(f"Items in TRAIN: {elms_in_train}")
                f.write(f"Items in TRAIN: {elms_in_train}\n")
                print(f"Items in TEST: {elms_in_test}")
                f.write(f"Items in TEST: {elms_in_test}\n")
            print("*"*10)
            f.write("*"*10)
            f.write("\n")
        print("*"*30)
        f.write("*"*30)
        f.write("\n")
```
